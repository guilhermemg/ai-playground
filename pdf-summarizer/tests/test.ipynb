```python
import requests

def make_request(target_url: str, pdf_file_path: str):
    with open(pdf_file_path, 'rb') as file:
        files = {"file": (file.name, file, "application/pdf")}
        response = requests.post(target_url, files=files)
    return response

# Example usage
target_url = "http://localhost:8002/summarize/"
pdf_file_path = "./data/resume.pdf"
response = make_request(target_url, pdf_file_path)

print(response)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Request failed with status code {response.status_code}")
```
